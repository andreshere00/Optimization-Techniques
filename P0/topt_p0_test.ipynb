```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

## Lectura de datos

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/TOPT/data/ppm_counts_per_min_squared.csv', sep=';')
df
```

```python
x=df[['ppm']]
y=df[['counts/min^2']]

plt.scatter(x, y)
plt.show()
```

## Modelo

```python
X_train, X_test, y_train, y_test = train_test_split(x,y)

clf = LinearRegression()
clf.fit(X_train, y_train)
```

## Representación de la solución

```python
m, b = np.polyfit(df['ppm'], df['counts/min^2'], 1)
```

```python
plt.plot(x, m*x + b, color='red')
plt.scatter(x, y)
plt.title('Gráfico de dispersión con línea de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Regression Line: y = {:.2f}x + {:.2f}'.format(m, b)])
```

## Predicción (interpolación) de nuevos datos

```python
predicted = clf.predict(df)
expected = y_test
```

## Validación del modelo

Lo haremos con el error cuadrático medio.

```python
sme_abs = np.sqrt(np.mean(np.array((expected-predicted)**2)))
sme_rel = np.sqrt(np.mean(np.array((expected-predicted)**2/expected)))
print(f'El error cuadrático medio absoluto es {round(sme_abs,2)}, normalizado a {round(sme_rel,2)}%')
```
